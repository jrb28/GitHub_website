First, let's do a very simple example of a matplotlib line chart

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [10000,20000, 30000, 25000]
plt.plot(x,y,label='Sales Revenue')   # 'Sales Revenue' is the label for the y data
plt.xlabel('Month')                   # Title for the horizontal axis
plt.ylabel('Sales Revenue')           # Title for the vertical axis
plt.axis([0,5,0,35000])               # sets the limits on the horizontal and vertical axes, respectively
plt.show()
```

This version is more aesthetically pleasing with month acronyms on the x axis labels using plt.xticks()

This code also has a statement to save the graphic as a .png file.  You must include this statement before the plt.show() statement

```python
x = [1,2,3,4]
x_labels = ['Jan','Feb','Mar','Apr']
y = [10000,20000, 30000, 25000]
plt.plot(x,y,label='Sales Revenue')   # 'Sales Revenue' is the label for the y data
plt.xlabel('Month')                   # Title for the horizontal axis
plt.ylabel('Sales Revenue')           # Title for the vertical axis
plt.xticks(x,x_labels)
plt.axis([0,5,0,35000])               # sets the limits on the horizontal and vertical axes, respectively
plt.savefig('sales.png')
plt.show()
```

Let's investigate the relationship of ozone level to wind speed

```python
""" Format of the scatterplot method is as follows: plt.scatter(x-series, y-series) """
plt.scatter(df_oz['wind'], df_oz['ozone'], alpha=0.5)
# alpha is a parameter that controls the transparency of the dots: 1 = solid, <1 = various transparency levels, 0 = no mark
plt.title('Ozone vs. Wind Speed')
plt.xlabel('Wind Speed')
plt.ylabel('Ozone')
plt.show()
```

```python
""" Format of the scatterplot method is as follows: plt.scatter(x-series, y-series) """
plt.scatter(df_oz['radiation'], df_oz['ozone'], alpha=0.5)
# alpha is a parameter that controls the transparency of the dots: 1 = solid, <1 = various transparency levels, 0 = no mark
plt.title('Ozone vs. Radiation')
plt.xlabel('Radiation (langleys)')
plt.ylabel('Ozone (ppb)')
plt.axis([0,400,0,200])
plt.show()
```

Code for comparing Smith and Green

```python
data = []       # create an empty Python list; sublists will be appended for each boxplot
data.append(list(df_test1.loc[df_test1['InstructorName'] == 'Smith']['EndYearTestScore']))
data.append(list(df_test1.loc[df_test1['InstructorName'] == 'Green']['EndYearTestScore']))
plt.boxplot(data)
plt.xticks([1,2],['Smith','Green'])
data_min = min([min(sublist) for sublist in data])  # this and the following 3 lines automatically size the graph to the data 
data_max = max([max(sublist) for sublist in data])  # while providing buffer space
plt.ylim([data_min - 2, data_max + 2]) 
plt.xlabel('Instructor')
plt.ylabel('Test Score')
plt.show()
```

Code for comparing Smith, Jones, and Green

```python
data = []       # create an empty Python list; sublists will be appended for each boxplot
instructors = ['Smith','Green','Jones']
for instructor in instructors:
    data.append(list(df_test1.loc[df_test1['InstructorName'] == instructor]['EndYearTestScore']))
plt.boxplot(data)
plt.xticks(range(1,len(instructors)+1),instructors)
data_min = min([min(sublist) for sublist in data])  # this and the following 3 lines automatically size the graph to the data 
data_max = max([max(sublist) for sublist in data])  # while providing buffer space
plt.ylim([data_min - 2, data_max + 2])   
plt.show()
```
